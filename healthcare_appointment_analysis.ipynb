### Overview
Patient no-shows represent a significant operational challenge for healthcare providers. This project conducts a comprehensive exploratory data analysis (EDA) on a dataset of 100,000 medical appointments in Brazil to identify the key factors that influence whether a patient shows up for their appointment. By understanding these drivers, healthcare systems can develop targeted interventions to improve attendance and optimize resource allocation.

### Technical Highlights
- **Data Wrangling:** Cleaning and preprocessing raw appointment data, including handling inconsistent entries and data type conversions.
- **Exploratory Data Analysis (EDA):** Investigating relationships between no-shows and patient demographics, scheduling patterns, and socio-economic factors.
- **Data Visualization:** Using Matplotlib and Seaborn to communicate findings through clear and informative charts.
- **Actionable Insights:** Deriving data-backed conclusions to inform healthcare operational strategies.

## 1. Data Loading and Cleaning
We begin by loading the appointment dataset and performing necessary cleaning steps to ensure data quality for our analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Load the dataset
df = pd.read_csv('noshowappointments-kagglev2-may-2016.csv')

# Initial Data Inspection
print(f"Dataset shape: {df.shape}")
display(df.head())

# Data Cleaning
# 1. Standardize column names
df.rename(columns={'No-show': 'No_show', 'Hipertension': 'Hypertension', 'Handcap': 'Handicap'}, inplace=True)

# 2. Convert date columns to datetime
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])

# 3. Remove illogical age values
df = df[df['Age'] >= 0]

# 4. Encode No_show as binary (1 for No-show, 0 for Show)
df['No_show_binary'] = df['No_show'].apply(lambda x: 1 if x == 'Yes' else 0)

print("Data cleaning complete.")
```

## 2. Exploratory Data Analysis
We explore various factors to understand their impact on patient attendance.

```python
# Overall No-Show Rate
no_show_rate = df['No_show_binary'].mean() * 100
print(f"Overall No-Show Rate: {no_show_rate:.2f}%")

# 1. Age and Attendance
plt.figure(figsize=(10, 6))
sns.boxplot(x='No_show', y='Age', data=df)
plt.title('Age Distribution by Attendance Status')
plt.show()

# 2. Scholarship and Attendance
plt.figure(figsize=(8, 5))
sns.barplot(x='Scholarship', y='No_show_binary', data=df)
plt.title('No-Show Rate by Scholarship Status')
plt.ylabel('No-Show Rate')
plt.show()

# 3. Day of the Week Analysis
df['DayOfWeek'] = df['AppointmentDay'].dt.day_name()
plt.figure(figsize=(10, 6))
order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
sns.barplot(x='DayOfWeek', y='No_show_binary', data=df, order=order)
plt.title('No-Show Rate by Day of the Week')
plt.ylabel('No-Show Rate')
plt.show()
```

## 3. Conclusions and Insights
Our analysis reveals several key findings:
- **Overall Rate:** Approximately 20% of appointments result in a no-show.
- **Age:** Younger patients are slightly more likely to miss appointments.
- **Scholarship:** Patients enrolled in the 'Scholarship' welfare program exhibit higher no-show rates.
- **Timing:** Attendance patterns vary by the day of the week, suggesting operational scheduling could be optimized.

These insights provide a foundation for developing targeted interventions, such as automated reminders for high-risk segments, to improve overall healthcare access and efficiency.
